```python
def tower_hanoi(n, a, b, c):
    if n == 1:
        print("move Ist disk from ", a, " to ", c)
        return
    
    tower_hanoi(n - 1, a, c, b)
    print("move nth disk from ", a, " to ", c)
    print()
    tower_hanoi(n - 1, b, a, c)
```
